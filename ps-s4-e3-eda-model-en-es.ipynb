```python
def show_corr_heatmap(df, title):
    
    corr = df.corr()
    mask = np.zeros_like(corr)
    mask[np.triu_indices_from(mask)] = True

    plt.figure(figsize = (15, 10))
    plt.title(title)
    # sns.heatmap(corr, annot = False, linewidths=.5, fmt=".2f", square=True, mask = mask, cmap=cmap_2)
    if df.shape[1] < 25:
        sns.heatmap(corr, annot=True, linewidths=.5, fmt=".2f", square=True, mask=mask, cmap=cmap_2)
    else:
        sns.heatmap(corr, annot=False, linewidths=.5, square=True, mask=mask, cmap=cmap_2)

    plt.show()
```

```python
def data_description(df):
    print("Data description")
    print(f"Total number of records {df.shape[0]}")
    print(f'number of features {df.shape[1]}\n\n')
    columns = df.columns
    data_type = []
    
    # Get the datatype of features
    for col in df.columns:
        data_type.append(df[col].dtype)
        
    n_uni = df.nunique()
    # Number of NaN values
    n_miss = df.isna().sum()
    
    names = list(zip(columns, data_type, n_uni, n_miss))
    variable_desc = pd.DataFrame(names, columns=["Name","Type","Unique levels","Missing"])
    print(variable_desc)
```

```python
def xgb_class(dataframe = train_df, num_fold = NUM_FOLD,
              shuffle = True, seed = SEED, objective=OBJECTIVE,
              eval_metric=EVAL_MET, target=TARGET):
    
    X = train_df.copy().drop(target, axis=1)
    y = train_df.copy()[target]
    
    cv = StratifiedKFold(n_splits=num_fold,shuffle=shuffle,random_state=seed)
    model = xgb.XGBClassifier(seed=seed,
                              objective=objective,
                              eval_metric=eval_metric)
    
    pipeline = Pipeline([
        ('classifier', model )
    ])
    
    auc_scores = []
    for train_idx, val_idx in cv.split(X, y.iloc[:,0]):
        X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
        y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

        eval_set = [(X_val, y_val)]
        pipeline.fit(X_train, y_train)

        y_pred = model.predict_proba(X_val)

        for i in range(len(target)):
            y_score = roc_auc_score(y_val.iloc[:, i], y_pred[:, i])
            auc_scores.append(y_score)

        avg_auc = np.mean(auc_scores)
        std_auc = np.std(auc_scores)
    print(f"{red}Avarage AUC:{res} {blue}{avg_auc}{res}")
    return X_train, X_val, y_train, y_val
```
