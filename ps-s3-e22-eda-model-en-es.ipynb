## <p style="font-family:Consolas Mono; font-weight:normal; letter-spacing: 2px; color:#06D1C7; font-size:130%; text-align:left;padding: 0px; border-bottom: 3px solid #008F77">Data information</p>

**🟦EN**:

This dataset is used to predict whether a horse will survive or not based on their past medical conditions. The key variable is **"outcome,"** which is found in the data and has the following values:

- 1 = Survived
- 2 = Died
- 3 = Was Euthanized

## Attribute Information

1. **Surgery**: Indicates whether the horse had surgery or not.
   - 1 = Yes, it had surgery
   - 2 = It was treated without surgery

2. **Age**: Classifies horses into age groups.
   - 1 = Adult horse
   - 2 = Young (< 6 months)

3. **Hospital Number**: A unique numeric identifier for each horse's case.

4. **Rectal Temperature**: Measures the horse's body temperature in degrees Celsius. Normal is 37.8, but it can change due to various factors.

5. **Pulse**: Reflects the heart rate in beats per minute.
   - Normal for adults: 30 - 40
   - Athletic horses may have 20 - 25
   - Elevated heart rate can indicate issues.

6. **Respiratory Rate**: Measures breaths per minute. Normal range is 8 to 10.

7. **Temperature of Extremities**: Indicates peripheral circulation.
   - 1 = Normal
   - 2 = Warm
   - 3 = Cool
   - 4 = Cold

8. **Peripheral Pulse**: A subjective assessment of peripheral circulation.
   - 1 = Normal
   - 2 = Increased
   - 3 = Reduced
   - 4 = Absent

9. **Mucous Membranes**: A subjective measurement of color of mucous membranes.
   - 1 = Normal pink
   - 2 = Bright pink
   - 3 = Pale pink
   - 4 = Pale cyanotic
   - 5 = Bright red / injected
   - 6 = Dark cyanotic

10. **Capillary Refill Time**: Indicates circulation quality.
    - 1 = < 3 seconds
    - 2 = >= 3 seconds

11. **Pain Level**: A subjective judgment of the horse's pain.
    - 1 = Alert, no pain
    - 2 = Depressed
    - 3 = Intermittent mild pain
    - 4 = Intermittent severe pain
    - 5 = Continuous severe pain

12. **Peristalsis**: Reflects gut activity.
    - 1 = Hypermotile
    - 2 = Normal
    - 3 = Hypomotile
    - 4 = Absent

13. **Abdominal Distension**: An important parameter indicating pain.
    - 1 = None
    - 2 = Slight
    - 3 = Moderate
    - 4 = Severe

14. **Nasogastric Tube**: Describes gas in the stomach.
    - 1 = None
    - 2 = Slight
    - 3 = Significant

15. **Nasogastric Reflux**: Reflects fluid obstruction.
    - 1 = None
    - 2 = > 1 liter
    - 3 = < 1 liter

16. **Nasogastric Reflux PH**: Measures acidity on a linear scale from 0 to 14. Normal values are 3 to 4.

17. **Rectal Examination - Feces**: Describes the condition of feces.
    - 1 = Normal
    - 2 = Increased
    - 3 = Decreased
    - 4 = Absent

18. **Abdomen**: Describes the abdominal condition.
    - 1 = Normal
    - 2 = Other
    - 3 = Firm feces in the large intestine
    - 4 = Distended small intestine
    - 5 = Distended large intestine

19. **Packed Cell Volume**: Measures red blood cell concentration in the blood. Normal range is 30 to 50.

20. **Total Protein**: Reflects dehydration level. Normal values: 6-7.5 (g/dL).

21. **Abdominocentesis Appearance**: Describes fluid obtained from the abdominal cavity.
    - 1 = Clear
    - 2 = Cloudy
    - 3 = Serosanguinous

22. **Abdominocentesis Total Protein**: Reflects gut health based on protein levels.

23. **Surgical Lesion**: Retrospective information about whether the problem was surgical.
    - 1 = Yes
    - 2 = No

24. **Type of Lesion**: Describes details of the lesion, including site, type, subtype, and specific code.

25. **CP Data**: Indicates the presence of pathology data for the case.
    - 1 = Yes
    - 2 = No

This dataset is valuable for predicting horse survival based on these medical attributes. However, it is essential to address missing data effectively for accurate analysis.

Este conjunto de datos se utiliza para predecir si un caballo sobrevivirá o no en función de sus condiciones médicas pasadas. La variable clave es **"resultado"**, que se encuentra en los datos y tiene los siguientes valores:

## Información de los Atributos

1. **Cirugía**: Indica si el caballo fue sometido a cirugía o no.
   - 1 = Sí, se realizó cirugía
   - 2 = Fue tratado sin cirugía

2. **Edad**: Clasifica a los caballos en grupos de edad.
   - 1 = Caballo adulto
   - 2 = Joven (< 6 meses)

3. **Número de Hospital**: Un identificador numérico único para cada caso de caballo.

4. **Temperatura Rectal**: Mide la temperatura corporal del caballo en grados Celsius. La temperatura normal es 37.8, pero puede cambiar debido a varios factores.

5. **Pulso**: Refleja la frecuencia cardíaca en pulsos por minuto.
   - Rango normal para adultos: 30 - 40
   - Raro tener una frecuencia más baja que la normal, aunque los caballos atléticos pueden tener una frecuencia de 20-25
   - Los animales con lesiones dolorosas o en estado de shock circulatorio pueden tener una frecuencia cardíaca elevada.

6. **Frecuencia Respiratoria**: Mide las respiraciones por minuto. El rango normal es de 8 a 10.

7. **Temperatura de Extremidades**: Indica la circulación periférica.
   - 1 = Normal
   - 2 = Cálido
   - 3 = Fresco
   - 4 = Frío

8. **Pulso Periférico**: Una evaluación subjetiva de la circulación periférica.
   - 1 = Normal
   - 2 = Aumentado
   - 3 = Reducido
   - 4 = Ausente

9. **Membranas Mucosas**: Una medición subjetiva del color de las membranas mucosas.
   - 1 = Rosado normal
   - 2 = Rosado brillante
   - 3 = Rosado pálido
   - 4 = Cianótico pálido
   - 5 = Rojo brillante / inyectado
   - 6 = Cianótico oscuro

10. **Tiempo de Recarga Capilar**: Indica la calidad de la circulación.
    - 1 = < 3 segundos
    - 2 = >= 3 segundos

11. **Nivel de Dolor**: Una evaluación subjetiva del dolor del caballo.
    - 1 = Alerta, sin dolor
    - 2 = Deprimido
    - 3 = Dolor leve intermitente
    - 4 = Dolor severo intermitente
    - 5 = Dolor severo continuo

12. **Peristalsis**: Refleja la actividad del intestino del caballo.
    - 1 = Hipermóvil
    - 2 = Normal
    - 3 = Hipomóvil
    - 4 = Ausente

13. **Distensión Abdominal**: Un parámetro importante que indica dolor.
    - 1 = Ninguna
    - 2 = Ligera
    - 3 = Moderada
    - 4 = Severa

14. **Tubo Nasogástrico**: Describe la presencia de gas en el estómago.
    - 1 = Ninguno
    - 2 = Ligero
    - 3 = Significativo

15. **Reflujo Nasogástrico**: Refleja la obstrucción del fluido.
    - 1 = Ninguno
    - 2 = > 1 litro
    - 3 = < 1 litro

16. **PH del Reflujo Nasogástrico**: Mide la acidez en una escala lineal de 0 a 14. Los valores normales están en el rango de 3 a 4.

17. **Examen Rectal - Heces**: Describe la condición de las heces.
    - 1 = Normal
    - 2 = Aumentado
    - 3 = Disminuido
    - 4 = Ausente

18. **Abdomen**: Describe la condición abdominal.
    - 1 = Normal
    - 2 = Otro
    - 3 = Heces firmes en el intestino grueso
    - 4 = Intestino delgado distendido
    - 5 = Intestino grueso distendido

19. **Volumen de Células Empaquetadas**: Mide la concentración de glóbulos rojos en la sangre. El rango normal es de 30 a 50.

20. **Proteína Total**: Refleja el nivel de deshidratación. Valores normales: 6-7.5 (g/dL).

21. **Apariencia de la Abdominocentesis**: Describe el fluido obtenido de la cavidad abdominal.
    - 1 = Claro
    - 2 = Nublado
    - 3 = Serosanguinolento

22. **Proteína Total de la Abdominocentesis**: Refleja la salud intestinal según los niveles de proteína.

23. **Lesión Quirúrgica**: Información retrospectiva sobre si el problema fue quirúrgico.
    - 1 = Sí
    - 2 = No

24. **Tipo de Lesión**: Describe detalles de la lesión, incluyendo sitio, tipo, subtipo y código específico.

25. **Datos de Patología (CP Data)**: Indica la presencia de datos de patología para el caso.
    - 1 = Sí
    - 2 = No

````python
def show_corr_heatmap(df, title):
    
    corr = df.corr()
    mask = np.zeros_like(corr)
    mask[np.triu_indices_from(mask)] = True

    plt.figure(figsize = (15, 10))
    plt.title(title)
    # sns.heatmap(corr, annot = False, linewidths=.5, fmt=".2f", square=True, mask = mask, cmap=cmap_2)
    if df.shape[1] < 25:
        sns.heatmap(corr, annot=True, linewidths=.5, fmt=".2f", square=True, mask=mask, cmap=cmap_2)
    else:
        sns.heatmap(corr, annot=False, linewidths=.5, square=True, mask=mask, cmap=cmap_2)

    plt.show()
````

````python
def data_description(df):
    print("Data description")
    print(f"Total number of records {df.shape[0]}")
    print(f'number of features {df.shape[1]}\n\n')
    columns = df.columns
    data_type = []
    
    # Get the datatype of features
    for col in df.columns:
        data_type.append(df[col].dtype)
        
    n_uni = df.nunique()
    # Number of NaN values
    n_miss = df.isna().sum()
    
    names = list(zip(columns, data_type, n_uni, n_miss))
    variable_desc = pd.DataFrame(names, columns=["Name","Type","Unique levels","Missing"])
    print(variable_desc)
````

````python
enc_df = train_df.copy()

def label_encoder(categorial_vars = categorical_vars, df = None):
    df_encoded = df.copy()
    label_encoders = {}

    for col in categorical_vars:
        enc = LabelEncoder()
        df_encoded[col] = enc.fit_transform(df[col])
        label_encoders[col] = enc
    
    return df_encoded
````
